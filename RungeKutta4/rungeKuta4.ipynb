## Runge-Kuta 4 ##

$y_{n+1}=y_{n}+\frac{h}{6}(K_{1}+2K_{2}+2K_{3}+K_{4})$

$K_{1}=f(x_{n},y_{n})$ 

$K_{2}=f(x_{n}+\frac{1}{2}h,y_{n}+\frac{1}{2}hK_{1})$ 

$K_{3}=f(x_{n}+\frac{1}{2}h,y_{n}+\frac{1}{2}hK_{2})$ 

$K_{4}=f(x_{n}+h,y_{n}+hK_{3})$ 


```python
def rk4(f,x0,xn,y0,n):
    x=np.linspace(x0,xn,n+1)
    y=np.zeros(n+1)
    h=(xn-x0)/n
    y[0]=y0
    
    for i in range(0,n):
        k1=f(x[i],y[i])
        k2=f(x[i]+0.5*h,y[i]+0.5*h*k1)
        k3=f(x[i]+0.5*h,y[i]+0.5*h*k2)
        k4=f(x[i]+h,y[i]+h*k3)
        
        y[i+1]=y[i]+(h/6)*(k1+2*k2+2*k3+k4)
    return (x,y)
```

```python
f=lambda x,y: 2*x*y

Sol=rk4(f,1,1.5,1,5)
```

```python
t=0
theta0=20*np.pi/180
theta_prima=0

betha=0.25
gamma=5.0

u=np.array([theta0,theta_prima])

def f(t,u):
    f1=u[1]
    f2=-betha*u[1]-gamma*np.sin(u[0])
    return np.array([f1,f2])

tsol=[t]
th_sol=[u[0]]
th_pr_sol=[u[1]]
h=0.01
tfin=10

while t<tfin:
    k1 = f(t,u)
    k2 = f(t + h/2 , u + h*k1/2)
    k3 = f(t + h/2 , u + h*k2/2)
    k4 = f(t + h, u + h*k3)
    u = u + (k1 + 2*(k2 + k3) + k4)*h/6
    t = t + h
    th_sol.append(u[0])
    th_pr_sol.append(u[1])
    tsol.append(t)

```

```python
plt.plot(tsol,th_sol)
plt.plot(tsol,th_pr_sol)
```
